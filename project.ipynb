```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# 'r' අකුර යොදා path එක නිවැරදි කර ඇත
dataset_path = r'C:\Users\kavee\Downloads\Crop Disease Detection\PlantVillage'

# Data Preprocessing (Rescaling සහ Train/Test ලෙස දත්ත වෙන් කිරීම)
datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

# Training Data සකස් කිරීම
print("Training Data:")
train_data = datagen.flow_from_directory(
    dataset_path,
    target_size=(224, 224), # පින්තූර වල ප්‍රමාණය
    batch_size=32,          # එකවර model එකට දෙන පින්තූර ගණන
    class_mode='categorical',
    subset='training'
)

# Validation (Testing) Data සකස් කිරීම
print("\nValidation Data:")
val_data = datagen.flow_from_directory(
    dataset_path,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# ඔයාගේ Dataset එකේ තියෙන ලෙඩ වර්ග ගණන (Classes) ලබා ගැනීම
num_classes = train_data.num_classes

# Model එක සෑදීම
model = Sequential([
    # 1 වන Convolutional සහ Pooling Layer එක
    Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    MaxPooling2D(pool_size=(2, 2)),

    # 2 වන Convolutional සහ Pooling Layer එක
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),

    # 3 වන Convolutional සහ Pooling Layer එක
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),

    # පින්තූර දත්ත 1D Array එකක් (දිගු පේළියක්) බවට පත් කිරීම
    Flatten(),

    # Fully Connected Layer එක
    Dense(128, activation='relu'),
    # Overfitting (model එක දත්ත කටපාඩම් කිරීම) අඩු කිරීමට Dropout එකක් දැමීම
    Dropout(0.5), 

    # Output Layer එක (රෝගය කුමක්ද යන්න සම්භාවිතාවයක් ලෙස දීමට)
    Dense(num_classes, activation='softmax')
])

# අපි හදපු Model එකේ සාරාංශයක් බලාගැනීමට
model.summary()
```

```python
# Model එක Compile කිරීම
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

print("Training Started... (මෙයට සුළු වේලාවක් ගත විය හැක)")

# Model එක Train කිරීම
history = model.fit(
    train_data,
    epochs=5, # මුළු දත්ත ගොනුවම 10 වතාවක් model එක හරහා යවනවා
    validation_data=val_data
)

print("Training Completed!")
```

```python
# Model එක 'crop_disease_model.h5' නමින් save කිරීම
model.save('crop_disease_model.h5')
print("Model Saved Successfully!")
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image

# 1. ඔයාගේ පින්තූරයේ Path එක (නම සහ තියෙන තැන හරියටම දෙන්න)
img_path = r'C:\Users\kavee\Downloads\Crop Disease Detection\PlantVillage\Tomato_healthy\ec57d9f8-7dcc-49e6-9ce7-af7d01eda4d1___RS_HL 9718.JPG' 

# 2. පින්තූරය Load කිරීම සහ 224x224 ප්‍රමාණයට වෙනස් කිරීම
img = image.load_img(img_path, target_size=(224, 224))

# 3. පින්තූරය Array එකක් (ඉලක්කම් ගොඩක්) බවට පත් කිරීම
img_array = image.img_to_array(img)

# 4. Rescale කිරීම (0 සහ 1 අතර අගයන් වලට)
img_array = img_array / 255.0

# 5. Model එකට දෙන්න පුළුවන් විදිහට shape එක හදාගැනීම
img_array = np.expand_dims(img_array, axis=0)

# 6. Model එකෙන් Prediction එක (අනුමානය) ලබා ගැනීම
predictions = model.predict(img_array)

# 7. වැඩිම සම්භාවිතාවය තියෙන ලෙඩේ මොකක්ද කියලා හොයාගැනීම
predicted_class_index = np.argmax(predictions)

# 8. Dataset එකේ තිබුණු Classes (ලෙඩ වර්ග) වල නම් ටික ලබාගැනීම
class_names = list(train_data.class_indices.keys())

# 9. අවසන් ප්‍රතිඵලය පෙන්වීම
print("\n" + "="*50)
print("🎯 මේ කොළයේ තියෙන රෝගය තමයි:", class_names[predicted_class_index])
print("📊 නිවැරදි වීමේ සම්භාවිතාවය:", round(np.max(predictions) * 100, 2), "%")
print("="*50 + "\n")
```

```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model

print("MobileNetV2 Base Model එක Download වෙමින් පවතී...")

# 1. ගූගල් සමාගමේ MobileNetV2 මොඩලය (Base Model) ගෙන ඒම
# include_top=False කියන්නේ අපි ඒකේ අවසාන කොටස (1000 classes) අයින් කරනවා කියන එකයි
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# 2. Base Model එකේ දැනුම වෙනස් නොවී තබා ගැනීම (Freeze කිරීම)
base_model.trainable = False

# 3. අපේ රෝග අඳුරගන්න අවශ්‍ය අලුත් Layers ටික එකතු කිරීම
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation='relu')(x)

# අපේ ලෙඩ වර්ග ගණනට සමානව අවසාන Layer එක හැදීම (train_data.num_classes මගින් ගණන ස්වයංක්‍රීයව ගනී)
predictions = Dense(train_data.num_classes, activation='softmax')(x)

# 4. සම්පූර්ණ Model එක එකලස් කිරීම
model = Model(inputs=base_model.input, outputs=predictions)

# 5. Model එක Compile කිරීම
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

print("අලුත් Model එක සාර්ථකව සැකසුවා! දැන් Train කිරීමට සූදානම්.")
```

```python
print("Training Started... (මෙයට සුළු වේලාවක් ගත විය හැක)")

history = model.fit(
    train_data,
    epochs=5, # මේ පාර 5ක් දුන්නම ඇති, මොකද එයා කලින්ම ගොඩක් දේවල් දන්නවා
    validation_data=val_data
)

print("Training Completed!")
```

```python
# ඔයාගේ අලුත්, දියුණු කළ Model එක Save කිරීම
model.save('super_crop_disease_model.keras')

print("සුපිරි Model එක සාර්ථකව පරිගණකයේ Save කළා! 🚀")
```

```python
import streamlit as st
import tensorflow as tf
from PIL import Image
import numpy as np

# වෙබ් පිටුවේ මාතෘකාව සහ හැඳින්වීම
st.set_page_config(page_title="Crop Disease Tracker", page_icon="🌱")
st.title("🌱 කෘෂිකාර්මික බෝග රෝග හඳුනාගැනීම")
st.write("ඔබේ වගාවේ රෝගී වූ කොළයක පින්තූරයක් මෙතනට Upload කරන්න.")

# Model එක Load කරගැනීම (අපි අලුතින් හදපු keras file එක)
@st.cache_resource
def load_model():
    model = tf.keras.models.load_model('super_crop_disease_model.keras')
    return model

model = load_model()

# පින්තූරය Upload කිරීමට අවකාශය හැදීම
uploaded_file = st.file_uploader("පින්තූරය තෝරන්න (JPG/PNG)...", type=["jpg", "jpeg", "png"])

if uploaded_file is not None:
    # Upload කරපු පින්තූරය වෙබ් එකේ පෙන්වීම
    image = Image.open(uploaded_file)
    st.image(image, caption='ඔබ Upload කළ පින්තූරය', use_container_width=True)
    st.write("පරීක්ෂා කරමින් පවතී... 🔍")

    # පින්තූරය Model එකට අවශ්‍ය විදිහට සකස් කිරීම
    image = image.convert('RGB') 
    img = image.resize((224, 224))
    img_array = tf.keras.preprocessing.image.img_to_array(img)
    img_array = img_array / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    # Prediction එක ලබා ගැනීම
    predictions = model.predict(img_array)
    predicted_class_index = np.argmax(predictions)
    confidence = round(np.max(predictions) * 100, 2)

    # ඔයාගේ Dataset එකේ තිබුණු රෝග 15 හි නම් ලැයිස්තුව
    class_names = [
        'Pepper__bell___Bacterial_spot', 'Pepper__bell___healthy', 
        'Potato___Early_blight', 'Potato___Late_blight', 'Potato___healthy', 
        'Tomato_Bacterial_spot', 'Tomato_Early_blight', 'Tomato_Late_blight', 
        'Tomato_Leaf_Mold', 'Tomato_Septoria_leaf_spot', 
        'Tomato_Spider_mites_Two_spotted_spider_mite', 'Tomato__Target_Spot', 
        'Tomato__Tomato_YellowLeaf__Curl_Virus', 'Tomato__Tomato_mosaic_virus', 
        'Tomato_healthy'
    ] 

    # ප්‍රතිඵලය ලස්සනට වෙබ් එකේ පෙන්වීම
    st.success(f"**හඳුනාගත් තත්වය:** {class_names[predicted_class_index]}")
    st.info(f"**නිවැරදි වීමේ සම්භාවිතාවය (Confidence):** {confidence}%")
```

```python
import tensorflow as tf

# 1. ඔයාගේ දැනට තියෙන මොඩල් එක Load කරගන්න 
# (ඔයාගේ ෆයිල් එකේ නම .keras ද .h5 ද කියලා බලලා හරියටම දෙන්න)
model = tf.keras.models.load_model('super_crop_disease_model.keras') 

# 2. TFLite Converter එක සකස් කිරීම
converter = tf.lite.TFLiteConverter.from_keras_model(model)

# 3. ෆෝන් එකට ගැළපෙන්න මොඩල් එක සැහැල්ලු කිරීම (Optimization - Optional but recommended)
converter.optimizations = [tf.lite.Optimize.DEFAULT]

# 4. මොඩල් එක TFLite බවට පත් කිරීම
tflite_model = converter.convert()

# 5. හැදුණු අලුත් TFLite ෆයිල් එක Save කිරීම
with open('crop_disease_model.tflite', 'wb') as f:
    f.write(tflite_model)

print("✅ TFLite Model එක සාර්ථකව හැදුවා! දැන් Download කරගන්න පුළුවන්.")
```
